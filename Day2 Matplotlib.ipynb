### 1) Line Plot : plt.plot()
##### When to Use:
- Continuous data (usually over time).
##### Data:
- X-axis: Represents the independent variable (e.g., time, categories).
- Y-axis: Represents the dependent variable (e.g., temperature, stock price).
##### What It Denotes:
- A trend over time (or another continuous variable).
- Shows how the value of a variable changes over time or some other continuous scale.
##### Example Use Case:
- Showing monthly sales over a year.
- Plotting temperature changes over the day.

```python
x = [1, 2, 3, 4]
y = [20, 40, 60, 80]
```

```python
# Create the line plot using all parameters
plt.plot(x, y,                    # creates a line plot
         #color='blue',            # Specifies the color of the line : 'blue', 'green' / color code ==> #bdf02b
         color='#bd1a91' ,
         linestyle='--',           # Defines the style of the line.  Options:'-','--',':','-.'
         linewidth=2,             # Sets the thickness of the line.  Higher values increase the thickness, and lower values make the line thinner.
         marker='o',              # Highlight datapoint using marking.  Options:'o': Circle,'^': Triangle up,'s': Square,'d': Diamond
         markersize=8,            # Marker size :Define size of marking
         markeredgecolor='red',   # Sets the color of the edge of the markers
         markerfacecolor='yellow',# Sets the fill color of the markers
         alpha=0.9,               # Define transparency of the line.  Range: 0 (completely transparent) to 1 (completely opaque).
         label='Line 1')          # Adds a label to the line, which can be displayed in the legend.

# Adding X-axis label
plt.xlabel("X-axis",              # Adds a label to the x-axis.
           fontsize=12,           # Specifies the font size of the label (default is 10).
           color='green',         # Sets the color of the label text (default is 'black').
           fontweight='bold',     # bold or normal.
           labelpad=10)           # Adds space between the axis and the label (default is 4).


# Adding Y-axis label
plt.ylabel("Y-axis", fontsize=12, color='green', labelpad=10,fontweight='bold' )  # Same as X label

# Adding title
plt.title("Line Plot Example",  # Adds a title to the plot.
          fontsize=14,          # Sets the size of the title text.
          color='purple')       # Sets the title text color.

# Add grid lines
plt.grid(True,                 # True/False: True means show the grid, False means hide it.
         which='minor',         # Specifies which grid lines to show. Options='both','major','minor'
         axis='both',          # Specifies which axis to apply the grid lines to. Options='both','x','y'
         linestyle='--',       # Defines the line style of the grid lines
         color='red',         # Define color of the grid lines
         alpha=0.8)            # Transparency level for the grid lines.


# Displaying legend (only for labeled elements)
plt.legend(loc='best')           # loc='best': Automatically selects the best position for the legend (avoiding overlaps with data).
                                 # Other positions Options: 'upper left', 'upper right', 'lower left', 'lower right', etc.

# Show the plot
plt.show()
```

```python
# Define figure size
plt.figure(figsize=(5, 2))  # Set figure size (width=10 inches, height=8 inches)

# Create the line plot using all parameters
plt.plot(x, y,                    # creates a line plot
         #color='blue',            # Specifies the color of the line : 'blue', 'green' / color code ==> #bdf02b
         color='#bd1a91' ,
         linestyle='--',           # Defines the style of the line.  Options:'-','--',':','-.'
         linewidth=2,             # Sets the thickness of the line.  Higher values increase the thickness, and lower values make the line thinner.
         marker='o',              # Highlight datapoint using marking.  Options:'o': Circle,'^': Triangle up,'s': Square,'d': Diamond
         markersize=8,            # Marker size :Define size of marking
         markeredgecolor='red',   # Sets the color of the edge of the markers
         markerfacecolor='yellow',# Sets the fill color of the markers
         alpha=0.9,               # Define transparency of the line.  Range: 0 (completely transparent) to 1 (completely opaque).
         label='Line 1')          # Adds a label to the line, which can be displayed in the legend.

# Adding X-axis label
plt.xlabel("X-axis",              # Adds a label to the x-axis.
           fontsize=12,           # Specifies the font size of the label (default is 10).
           color='green',         # Sets the color of the label text (default is 'black').
           fontweight='bold',     # bold or normal.
           labelpad=10)           # Adds space between the axis and the label (default is 4).


# Adding Y-axis label
plt.ylabel("Y-axis", fontsize=12, color='green', labelpad=10,fontweight='bold' )  # Same as X label

# Adding title
plt.title("Line Plot Example",  # Adds a title to the plot.
          fontsize=14,          # Sets the size of the title text.
          color='purple')       # Sets the title text color.

# Add grid lines
plt.grid(True,                 # True/False: True means show the grid, False means hide it.
         which='minor',         # Specifies which grid lines to show. Options='both','major','minor'
         axis='both',          # Specifies which axis to apply the grid lines to. Options='both','x','y'
         linestyle='--',       # Defines the line style of the grid lines
         color='red',         # Define color of the grid lines
         alpha=0.8)            # Transparency level for the grid lines.


# Displaying legend (only for labeled elements)
plt.legend(loc='best')           # loc='best': Automatically selects the best position for the legend (avoiding overlaps with data).
                                 # Other positions Options: 'upper left', 'upper right', 'lower left', 'lower right', etc.

# Show the plot
plt.show()
```

### 2) Bar Chart :plt.bar()
#### When to Use:
- Categorical data.
When you want to compare quantities across different categories.
#### Data:
- X-axis: Categories or groups.
- Y-axis: Values or counts associated with the categories.
#### What It Denotes:
- The comparisons between different categories.
- The size of each bar represents the magnitude of each category.
#### Example Use Case:
- Comparing sales by product category.
- Visualizing survey responses (e.g., "Agree", "Neutral", "Disagree").

```python
Subject = ['Python', 'MySQL', 'Power BI', 'Excel']
Number_Student = [20, 25, 10, 15]
```

```python
plt.bar(Subject,Number_Student)
plt.xlabel("Subject")
plt.ylabel("Number of Student")
plt.title("Enrolment of students")
plt.show()
```

```python
plt.barh(Subject,Number_Student)
plt.xlabel("Subject")
plt.ylabel("Number of Student")
plt.title("Enrolment of students")
plt.show()
```

```python
# Define figure size
plt.figure(figsize=(10, 8))  # Set figure size (width=10 inches, height=8 inches)

bars=plt.bar(
    x=Subject,                         # X-axis values (categories)
    height=Number_Student,             # Y-axis values (numeric data)
    color=['blue', 'green', 'red', 'purple'], # Bar colors
    edgecolor='black',                 # Border color of bars
    linewidth=1.5,                      # Width of the border line
    alpha=0.7,                          # Transparency of bars (0: fully transparent, 1: opaque)
    width=0.5,                          # Width of bars (default is 0.8)
    align='center'                      # Alignment of bars ('center', 'edge')
)


plt.xlabel("Subjects", fontsize=12, color='blue')  # X-axis label
plt.ylabel("Number of Students", fontsize=12, color='green')  # Y-axis label
plt.title("Enrolment of Students by Subject", fontsize=14,  color='red')  # Chart title


plt.grid(axis='x', linestyle='--', alpha=0.8)  # grid lines

plt.legend(bars, Subject, title="Subjects", loc="upper right")   # legend

plt.show()
```
