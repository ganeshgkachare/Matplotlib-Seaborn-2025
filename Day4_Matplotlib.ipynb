### 4) Scatter Plot :plt.scatter()
#### When to Use:
- To identify Correlation or relationship between two variables.

#### Data:
- X-axis: Represents one variable.
- Y-axis: Represents the other variable.
#### What It Denotes:
- A visual represents relationships between two variables
- Helps identify correlations, trends
- to determine if there is a correlation (positive, negative, or none) between variables.
#### Example Use Case:
- Examining the relationship between hours studied and exam scores.
- Plotting age vs. income of a population.

```python
import matplotlib.pyplot as plt
```

```python
Experience = [1, 2, 3, 4, 5]
CTC = [4, 6, 9, 12, 15]
```

```python
plt.scatter(
    Experience, CTC, 
    s=90,  # Size of markers
    color='red',  # Color of markers
    marker='o',  # Marker style (circle)
    alpha=0.7,  # Transparency (0: fully transparent, 1: opaque)
    edgecolors='black',  # Edge color of markers
    linewidths=1.5,  # Width of marker edges
)


plt.xlabel("Experience in Years")  # Label for x-axis
plt.ylabel("CTC in Lakhs")  # Label for y-axis
plt.title("Relationship between Experience and CTC")  # Title of the plot

# Add grid lines
plt.grid(True,                 # True/False: True means show the grid, False means hide it.
         axis='both',          # Specifies which axis to apply the grid lines to. Options='both','x','y'
         linestyle='--',       # Defines the line style of the grid lines
         color='green',         # Define color of the grid lines
         alpha=0.5)            # Transparency level for the grid lines.


# Display the plot
plt.show()
```

```python
Experience = [1, 2, 3, 4, 5]
CTC = [4, 6, 9, 12, 15]

Experience1 = [1, 1.5, 2.8, 3.5, 4.5]
CTC1 = [3, 5, 8, 11, 13]
```

```python
plt.scatter(
    Experience, CTC, 
    s=100,  # Size of markers
    c='#397f20',  # Color of markers
    marker='o',  # Marker style (circle)
    alpha=0.7,  # Transparency (0: fully transparent, 1: opaque)
    edgecolors='black',  # Edge color of markers
    linewidths=1.5,  # Width of marker edges
    label="Male"  # Legend label
)

plt.scatter(
    Experience1, CTC1, 
    s=100,  # Size of markers
    c='#22e7ed',  # Color of markers
    marker='s',  # Marker style (circle)
    alpha=0.7,  # Transparency (0: fully transparent, 1: opaque)
    edgecolors='black',  # Edge color of markers
    linewidths=1.5,  # Width of marker edges
    label="Female"  # Legend label
)

plt.xlabel("Experience in Years")  # Label for x-axis
plt.ylabel("CTC in Lakhs")  # Label for y-axis
plt.title("Relationship between Experience and CTC")  # Title of the plot
plt.grid(True, linestyle="--", alpha=0.5) # Grid lines


plt.legend() # Legend

# Display the plot
plt.show()
```

### 6) Pie Chart : plt.pie()
#### When to Use:
- To show contribution of each category.
#### Data:
- category/ segment
- segment numeric value
#### What It Denotes:
- The proportion of each category in relation to the total.
#### Example Use Case:
- Visualizing market share by company.
- Showing the distribution of expenses in a budget.

```python
departments = ['Marketing', 'HR', 'Operations', 'IT', 'Admin', "R&D"] 
Budget = [8, 7, 5, 14, 6, 10]  
```

```python
plt.figure(figsize=(10,10))
plt.pie(
    Budget,              # Input data numeric
    labels=departments, # Input data category
    autopct='%1.2f%%',  # Display percentage values /d
    colors=['gold', 'skyblue', 'lightcoral', 'limegreen', 'orange',"green"],  # Assign colors to slices
    # explode=(0.2, 0, 0, 0.5, 0,0),  # Highlight the first slice
    explode=(0, 0, 0, 0.5, 0,0),
    startangle=220,  # Rotate for better visualization
    shadow=False,  # Add a shadow effect
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.0},  # Border for each slice
    pctdistance=0.8,  # Position of percentage labels
    labeldistance=1.1  # Position of department labels
)


plt.title("Company Budget Distribution by Department")

plt.show()
```

```python
# Sample data
labels = ['A', 'B', 'C']
values = [35, 45, 25]
plt.figure(figsize=(5,5))
# Create the pie chart
plt.pie(values, labels=labels, autopct='%1.2f%%')
# plt.pie(values, labels=labels)

# Add title
plt.title('Pie Chart')
plt.legend(labels, loc='upper right') # upper left, lower left, lower right
# Display the graph
plt.show()
```
