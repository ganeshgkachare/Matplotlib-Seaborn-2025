#### Common pyplot Functions

| Function                | Description                        |
|-------------------------|----------------------------------|
| `plt.plot(x, y)`        | Creates a line plot             |
| `plt.scatter(x, y)`     | Creates a scatter plot          |
| `plt.bar(x, height)`    | Creates a bar chart            |
| `plt.hist(data, bins)`  | Creates a histogram            |
| `plt.boxplot(data)`     | Creates a box plot             |
| `plt.pie(sizes, labels)`| Creates a pie chart            |
| `plt.xlabel("X-axis")`  | Adds label to X-axis           |
| `plt.ylabel("Y-axis")`  | Adds label to Y-axis           |
| `plt.title("Title")`    | Adds title to the plot         |
| `plt.legend()`          | Adds a legend                  |
| `plt.grid(True)`        | Adds grid lines                |
| `plt.savefig("plot.png")` | Saves the plot as an image file |


### 1) Line Plot : plt.plot()
##### When to Use:
- Continuous data (usually over time).
##### Data:
- X-axis: Represents the independent variable (e.g., time, categories).
- Y-axis: Represents the dependent variable (e.g., temperature, stock price).
##### What It Denotes:
- A trend over time (or another continuous variable).
- Shows how the value of a variable changes over time or some other continuous scale.
##### Example Use Case:
- Showing monthly sales over a year.
- Plotting temperature changes over the day.

```python
x = [1, 2, 3, 4]
y = [20, 40, 60, 80]
```

```python
plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot Example")
plt.grid(True)
plt.show()
```

```python
# Create the line plot using all parameters
plt.plot(x, y,                    # creates a line plot
         #color='blue',            # Specifies the color of the line : 'blue', 'green' / color code ==> #bdf02b
         color='#bd1a91' ,
         linestyle='--',           # Defines the style of the line.  Options:'-','--',':','-.'
         linewidth=2,             # Sets the thickness of the line.  Higher values increase the thickness, and lower values make the line thinner.
         marker='o',              # Highlight datapoint using marking.  Options:'o': Circle,'^': Triangle up,'s': Square,'d': Diamond
         markersize=8,            # Marker size :Define size of marking
         markeredgecolor='red',   # Sets the color of the edge of the markers
         markerfacecolor='yellow',# Sets the fill color of the markers
         alpha=0.7,               # Define transparency of the line.  Range: 0 (completely transparent) to 1 (completely opaque).
         label='Line 1')          # Adds a label to the line, which can be displayed in the legend.

# Adding X-axis label
plt.xlabel("X-axis",              # Adds a label to the x-axis.
           fontsize=12,           # Specifies the font size of the label (default is 10).
           color='green',         # Sets the color of the label text (default is 'black').
           labelpad=10)           # Adds space between the axis and the label (default is 4).


# Adding Y-axis label
plt.ylabel("Y-axis", fontsize=12, color='green', labelpad=10)  # Same as X label

# Adding title
plt.title("Line Plot Example",  # Adds a title to the plot.
          fontsize=14,          # Sets the size of the title text.
          color='purple')       # Sets the title text color.

# Add grid lines
plt.grid(True,                 # True/False: True means show the grid, False means hide it.
         which='both',         # Specifies which grid lines to show. Options='both','major','minor'
         axis='both',          # Specifies which axis to apply the grid lines to. Options='both','x','y'
         linestyle='--',       # Defines the line style of the grid lines
         color='gray',         # Define color of the grid lines
         alpha=0.5)            # Transparency level for the grid lines.


# Displaying legend (only for labeled elements)
plt.legend(loc='best')           # loc='best': Automatically selects the best position for the legend (avoiding overlaps with data).
                                 # Other positions Options: 'upper left', 'upper right', 'lower left', 'lower right', etc.

# Show the plot
plt.show()
```
