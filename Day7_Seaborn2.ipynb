```python
# Import seaborn
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
# Scatter Plot 
# Correlation Between Two Variables
# Used to identify relationships between numerical variables.


sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Total Bill vs. Tip Amount")
plt.show()
```

```python

sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips)
plt.title("Total Bill vs. Tip Amount")
plt.show()
```

```python

sns.scatterplot(x="total_bill", y="tip", hue="size",data=tips)
plt.title("Total Bill vs. Tip Amount")
plt.show()
```

```python

# Line plot
plt.figure(figsize=(12, 6))

sns.lineplot(data=Sale, x="Month", y="Sales")

plt.title("Monthly Sales", fontsize=14)

plt.show()
```

```python
Sale.groupby("Month")['Sales'].sum().reset_index()
```

```python
data2=Sale.groupby("Month")['Sales'].sum().reset_index()
data2.head()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x="Month", y='Sales', marker="o",data=data2)

# Customizing the plot
plt.title("Monthly Sales Trend", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x="Month", y='Sales',hue="Region", marker="o",data=Sale)

# Customizing the plot
plt.title("Monthly Sales Trend", fontsize=14)

plt.show()
```
