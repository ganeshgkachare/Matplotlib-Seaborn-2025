### 3) Histogram : plt.hist()
#### When to Use:
- Distribution of a continuous dataset.
- When you want to see the frequency of data points within specific ranges (bins).
#### Data:
- X-axis: Continuous data.
- Y-axis: Frequency of values within each bin (or range).
#### What It Denotes:
- The distribution of values in a dataset.
- How frequently values fall into different ranges.
#### Example Use Case:
- Analyzing the distribution of test scores.
- Visualizing the distribution of a dataset like ages or income.

```python
Marks = [82,73,42,21,52,22,75,67,67,21,54,60,43,77,94,75,42,61,83,52,77,72,87,50,56,50,76]
```

```python
plt.hist(Marks,  # Data
    bins=5,     #  Number of bins (interval)
    color='#262bea', 
    edgecolor='black', 
    linewidth=1.5, 
    alpha=0.7, 
    histtype='bar', 
    rwidth=0.5
)

plt.xlabel("Marks", fontsize=12, fontweight='bold', color='blue')
plt.ylabel("Number of students", fontsize=12, fontweight='bold', color='blue')
plt.title("Histogram", fontsize=14, fontweight='bold', color='darkred')
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adds horizontal grid lines
plt.show()
```

```python
# A stacked histogram is a type of histogram where multiple datasets are stacked on top of each other in the same bins, instead of overlapping.

Marks_Male = [82,73,42,21,52,22,75,67,36,67,21,54,60,43,77,94,75,92,61,83,52,77,72,87,50,56,50,76]
Marks_Female = [80,73,42,21,52,22,75,35,37,67,21,54,40,43,77,94,75,42,61,83,52,77,72,87,50,56,50,76]
```

```python
plt.figure(figsize=(10, 8))  # Set figure size

plt.hist( [Marks_Male, Marks_Female],  # Datasets
    bins=4,                      # Number of bins (interval)
    color=['green', 'salmon'],  # Colors for each dataset
    edgecolor='black',        # Color of bar edges
    linewidth=1.2,         # Thickness of bar edges
    alpha=0.7,            # Transparency of bars
    histtype='bar',  # Histogram type : 'barstacked' /'step' /'stepfilled' /'bar'
    align='right',     # left / right
    rwidth=0.55,  # Relative width of bars
    stacked=True,  # Stack the bars
    label=['Boys', 'Girls ']  # Labels for the legend
)

# Add Axis name and Title
plt.xlabel("Marks", fontsize=12, color='green')
plt.ylabel("Number of students", fontsize=12,  color='green')
plt.title("Stacked Histogram ", fontsize=14,  color='black')


plt.legend(title="Groups", loc='upper left', fontsize=10, title_fontsize=11)  # Add Legend

plt.grid(axis='both', linestyle='--', alpha=0.7)  # Add Grid

plt.show()  # Show the Plot
```

### 4) Scatter Plot :plt.scatter()
#### When to Use:
- To identify Correlation or relationship between two variables.

#### Data:
- X-axis: Represents one variable.
- Y-axis: Represents the other variable.
#### What It Denotes:
- A visual represents relationships between two variables
- Helps identify correlations, trends
- to determine if there is a correlation (positive, negative, or none) between variables.
#### Example Use Case:
- Examining the relationship between hours studied and exam scores.
- Plotting age vs. income of a population.
